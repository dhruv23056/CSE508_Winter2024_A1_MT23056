**For printing a file **

```python
def Print(folder_path, num_files=5):
    file_list = [f"file{i}.txt" for i in range(1, num_files + 1)]

    for filename in file_list:
        file_path = os.path.join(folder_path, filename)

        try:
            with open(file_path, 'r') as file:
                content = file.read()
                print(f"Contents of {filename}:\n{content}\n")
        except FileNotFoundError:
            print(f"File {filename} not found.\n")
```

**remove HTML Tags**

```python
from bs4 import BeautifulSoup
import os

def remove_html_tags(text):
    soup = BeautifulSoup(text, "html.parser")
    return soup.get_text()

def process_files(input_directory):
    file_list = os.listdir(input_directory)

    for file_name in file_list:
        if file_name.endswith(".txt"):
            file_path = os.path.join(input_directory, file_name)

            with open(file_path, 'r', encoding='utf-8') as file:
                content = file.read()
                cleaned_content = remove_html_tags(content)

            with open(file_path, 'w', encoding='utf-8') as file:
                file.write(cleaned_content)

            print(f"Processed: {file_name}")

input_directory = "/content/drive/MyDrive/temp/text_files"

process_files(input_directory)
```

**3) StopWords**

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk
nltk.download('stopwords')


print("Before StopWords:\n")
Print('/content/drive/MyDrive/temp/text_files/')

def remove_stopwords(data):
    stop_words = set(stopwords.words("english"))
    tokens = word_tokenize(data)
    without_stopwords = [w for w in tokens if w.lower() not in stop_words]
    return without_stopwords

for file in dir_list:
    input_file_path = '/content/drive/MyDrive/temp/text_files/' + file

    with open(input_file_path, 'r') as input_file:
        lines = input_file.readlines()
        modified_lines = [" ".join(remove_stopwords(line)) for line in lines]

    with open(input_file_path, 'w') as output_file:
        for modified_line in modified_lines:
            output_file.write(modified_line + '\n')

print("After StopWords:\n")
Print('/content/drive/MyDrive/temp/text_files/')
```

**4) punctuations**

```python
import string

print("Before Punctuation:\n")
Print('/content/drive/MyDrive/temp/text_files/')


for file in dir_list:
    input_file_path = '/content/drive/MyDrive/temp/text_files/' + file

    with open(input_file_path, 'r') as input_file:
        data = input_file.read()

    translator = str.maketrans('', '', string.punctuation)
    data_without_punctuation = data.translate(translator)

    with open(input_file_path, 'w') as output_file:
        output_file.write(data_without_punctuation)

print("After Punctuaton:\n")
Print('/content/drive/MyDrive/temp/text_files/')
```

**5) blank space tokens**

```python
print("Before Blankspace:\n")
Print('/content/drive/MyDrive/temp/text_files/')


for file in dir_list:
    input_file_path = '/content/drive/MyDrive/temp/text_files/' + file

    with open(input_file_path, 'r') as input_file:
        lines = input_file.readlines()
        modified_lines = [" ".join(line.split()) for line in lines]

    with open(input_file_path, 'w') as output_file:
        for modified_line in modified_lines:
            output_file.write(modified_line + '\n')

print("After Blanckspace:\n")
Print('/content/drive/MyDrive/temp/text_files/')
```
